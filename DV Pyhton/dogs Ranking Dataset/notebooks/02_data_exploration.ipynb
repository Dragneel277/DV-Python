# Data Exploration

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load cleaned data
cleaned_data = pd.read_csv("../data/cleaned_dogs_data.csv")

# Step 1: Basic Statistics
print(cleaned_data.describe())

# Step 2: Visualizations

# Distribution of Longevity
plt.figure(figsize=(10, 6))
sns.histplot(cleaned_data['LONGEVITY(YEARS)'], kde=True)
plt.title('Distribution of Longevity (Years)')
plt.xlabel('Longevity (Years)')
plt.ylabel('Frequency')
plt.show()

# Scatter plot of Longevity vs. Suitability for Children
plt.figure(figsize=(10, 6))
sns.scatterplot(x='LONGEVITY(YEARS)', y='SUITABILITY FOR CHILDREN', data=cleaned_data)
plt.title('Longevity vs. Suitability for Children')
plt.xlabel('Longevity (Years)')
plt.ylabel('Suitability for Children')
plt.show()


# Assuming cleaned_data is your DataFrame
# Select only numeric columns for the correlation heatmap
numeric_data = cleaned_data.select_dtypes(include=[float, int])

# Generate the heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(numeric_data.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```
